```python
import json
import re
from collections import defaultdict
from itertools import permutations
```

```python
d={}
with open("data_wordmap.json","r",encoding="utf-8") as fp:
    d=json.load(fp)
```

```python
ekar=[]
for key in d:
    b=key
    if b.endswith('ি'):
        b=b[:-1]+'ী'
    else:
        continue
    if b in d:
        ekar.append([key,b])
```

```python
print(ekar)
len(ekar)
```

```python
all_a = []
a=[]
for key in d:
    b=key
    if b.startswith('অ্যা'):
        b="এ"+b[1:]
    else:
        continue
    if b in d:
        a.append([key,b])
        all_a.append(key)
        all_a.append(b)

print(a)
len(a)
```

```python
a_2=[]
for key in d:
    b=key
    if "অ্যা" in b:
        b=re.sub("অ্যা", "এ", b)
    else:
        continue
    if b in d and (key not in all_a or b not in all_a):
        a_2.append([key,b])
        all_a.append(key)
        all_a.append(b)

a_2
```

```python
a_3=[]
for key in d:
    b=key
    if "এ্যা" in b:
        b1=re.sub("এ্যা", "অ্যা", b)
        b2=re.sub("এ্যা", "এ", b)
    else:
        continue

    if b1 in d and (key not in all_a or b1 not in all_a):
        a_3.append([key,b1])
        all_a.append(key)
        all_a.append(b1)
    if b2 in d and (key not in all_a or b2 not in all_a):
        a_3.append([key,b2])
        all_a.append(key)
        all_a.append(b2)
a_3
```

```python
all_a = [[all_a[i],all_a[i+1]] for i in range(0,len(all_a),2)]
all_a
```

```python
nt = []
for key in d:
    b=key
    if "ন্ট" in b:
        b=re.sub("ন্ট", "ণ্ট", b)
    else:
        continue
    if b in d:
        nt.append([key,b])
nt
```

```python
hosonto=[]
for key in d:
    b=key
    if "্" in b:
        b=re.sub("্", "", b)
    else:
        continue
    if b in d:
        hosonto.append([key,b])

print("Total",len(hosonto))
print(hosonto)
```

```python
dontonya=[]
for key in d:
    b=key
    if "ন" in b:
        b=re.sub("ন", "ণ", b)
    else:
        continue
    if b in d:
        dontonya.append([key,b])

print("Total",len(dontonya))
print(dontonya)
```

```python
jofola=[]
for key in d:
    b=key
    if "্যা" in b:
        b=re.sub("্যা", "া", b)
    elif "যা" in b:
        b=re.sub("যা", "া", b)
    else:
        continue
    if b in d:
        jofola.append([key,b])

print("Total",len(jofola))
print(jofola)
```

```python
nkh=[]
for key in d:
    b=key
    if "ঙ্খ" in b:
        b=re.sub("ঙ্খ", "ঙ্ক্ষ", b)
    elif "ঙ্" in b:
        b=re.sub("ঙ্", "ং", b)
    elif "ঙ" in b:
        b=re.sub("ঙ", "ং", b)
    else:
        continue
    if b in d:
        nkh.append([key,b])

print("Total",len(nkh))
print(nkh)
```

```python
chondobindu = []
for key in d:
    b=key
    if "ঁ" in b:
        b=re.sub("ঁ", "", b)
    else:
        continue
    if b in d:
        chondobindu.append([key,b])

print("Total",len(chondobindu))
print(chondobindu)
```

```python
ksh=[]
for key in d:
    b=key
    if "কস" in b:
        b=re.sub("কস", "ক্স", b)
    else:
        continue
    if b in d:
        ksh.append([key,b])

print("Total",len(ksh))
print(ksh)
```

```python
previous_all = [*ekar, *all_a,*nt,*hosonto,*dontonya,*jofola,*nkh, *chondobindu, *ksh]
missed=[]

# combining logic variations
for key in d:
    b=key
    changed=False
    if b.startswith('অ্যা'):
        b="এ"+b[1:]
        changed=True
    if "অ্যা" in b:
        b=re.sub("অ্যা", "এ", b)
        changed=True
    if "্যা" in b:
        b=re.sub("্যা", "া", b)
        changed=True
    elif "যা" in b:
        b=re.sub("যা", "া", b)
        changed=True
    if "ন্ট" in b:
        b=re.sub("ন্ট", "ণ্ট", b)
        changed=True
    if "ঁ" in b:
        b=re.sub("ঁ", "", b)
        changed=True
    if b.endswith('ি'):
        b=b[:-1]+'ী'
        changed=True
    if "ঙ্খ" in b:
        b=re.sub("ঙ্খ", "ঙ্ক্ষ", b)
        changed=True
    elif "ঙ্" in b:
        b=re.sub("ঙ্", "ং", b)
        changed=True
    elif "ঙ" in b:
        b=re.sub("ঙ", "ং", b)
        changed=True
    if "্" in b:
        b=re.sub("্", "", b)
        changed=True
    if "কস" in b:
        b=re.sub("কস", "ক্স", b)
        changed=True
    if not changed:
        continue
    if b in d and b not in previous_all:
        missed.append([key,b])

print("Total",len(missed))
print(missed)
```

```python
all = [*missed, *previous_all]
s = [sorted(i) for i in all]
all_unique=defaultdict(set)
for items in s:
    all_unique[items[0]].add(items[1])
    
l=[]
for i in all:
    l.extend(i)

# print("Total unique words", len(set(l)))
# print(l)
```

```python
print("Unique keys",len(all_unique) )
# all_unique
```

```python
def get_variations(word, return_original=True):
    variations = set()
    idx=[match.start() for match in re.finditer(r"ী|ি", word)]
    len_idx=len(idx)
    combos=[i for i in permutations(idx)]
    for combo in combos:
        out = word
        for pos in combo:
            if out[pos]=="ী":
                out=out[:pos]+"ি"+out[pos+1:]
            elif out[pos]=="ি":
                out=out[:pos]+"ী"+out[pos+1:]
            variations.add(out)
    if return_original:
        variations.add(word)
    return variations

# set return_original=False, to get all combinations except input word
get_variations("সহধর্মিনী",return_original=False)
```

```python
additions=defaultdict(set)
v={}

for key,values_set in all_unique.items():
    variations = set()
    if "ি" in key or "ী" in key:
        variations.update(get_variations(key))
    else:
        continue
    
    for value in values_set:
        # if key has "ি" or "ী", then words in values_set will definately have them
        # so, no need to check again
        variations.update(get_variations(value))
    
    v[key]=variations

    for variation in variations:
        if variation in all_unique:
            # already included as key, no need to add to value
            continue

        if variation in d and variation not in values_set:
            all_unique[key].add(variation)
            additions[key].add(variation)

# updates added to all_unique dictionary
additions
```
